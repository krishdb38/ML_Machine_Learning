#### Machine Learning

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score , mean_squared_error
from sklearn.ensemble import RandomForestRegressor
```

```python
x = data.drop(["charges"] , axis = 1) # drop the column axis
y = data.charges
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, random_state = 0)
```

#### Linear Regression

```python
lr = LinearRegression().fit(x_train, y_train)
```

```python
y_train_pred = lr.predict(x_train)
y_test_pred = lr.predict(x_test)
```

```python
print(lr.score(x_test, y_test))
```

Linear Regression is also predicting very well. Lets apply other Methods
Polynomial signs and look the results

```python
x = data.drop(["charges", "region"], axis = 1) # Drop the column
y = data.charges

quad = PolynomialFeatures(degree = 2)
x_quad = quad.fit_transform(x)
```

```python
x_quad
```

```python
x_train , x_test, y_train , y_test = train_test_split(x_quad, y,random_state =0)
plr = LinearRegression().fit(x_train , y_train)

y_train_pred = plr.predict(x_train)
y_test_pred = plr.predict(x_test)

print(plr.score(x_test , y_test))
```

```python
# We can see the result is increasesd very high
# Lets use Random Forest Regression
```

```python
forest = RandomForestRegressor(n_estimators=100,
                              criterion= "mse",
                              random_state=1,
                              n_jobs = -1)
forest.fit(x_train , y_train)

forest_train_predict = forest.predict(x_train)
forest_test_predict = forest.predict(x_test)

print("MSE train data : %.3f, MSE test data: %.3f"%(
    mean_squared_error(y_train, forest_train_predict),
    mean_squared_error(y_test, forest_test_predict)))

print('R2 train data: %.3f, R2 test data: %.3f' % (
    r2_score(y_train,forest_train_predict),
    r2_score(y_test,forest_test_predict)))
```

```python
plt.figure(figsize=(10,6))
plt.scatter(forest_train_predict , forest_train_predict-y_train,
           c = "black", marker = "o", s = 35, alpha = 0.7,
           label = "Test data")
plt.xlabel("Predicted values")
plt.ylabel("Tailings")
plt.legend(loc = "upper left")
plt.hlines(y = 0, xmin = 0 , xmax = 60000, lw = 2 , color = "red")
plt.show()
```
