## 9.1 파이썬을 이용하여 CIFAR-10으로부터 이미지 불러오기

```python
import pickle
def unpickle(file):
    fo = open(file, 'rb')
    dict = pickle.load(fo, encoding='latin1')
    fo.close()
    return dict
```

## 9.2 데이터 정제

```python
import numpy as np
```

```python
def clean(data):
    imgs = data.reshape(data.shape[0], 3, 32, 32) # 3개의 채널을 가지는 32 x 32 행렬로 재구성해줍니다
    grayscale_imgs = imgs.mean(1) # 컬러의 강도를 평균해 줌으로써 이미지를 흑백으로 만들어줍니다
    cropped_imgs = grayscale_imgs[:, 4:28, 4:28] # 32 x 32 이미지를 24 x 24 이미지로 잘라내 줍니다
    img_data = cropped_imgs.reshape(data.shape[0], -1)
    img_size = np.shape(img_data)[1]
    means = np.mean(img_data, axis=1)
    meansT = means.reshape(len(means), 1)
    stds = np.std(img_data, axis=1)
    stdsT = stds.reshape(len(stds), 1)
    adj_stds = np.maximum(stdsT, 1.0 / np.sqrt(img_size))
    normalized = (img_data - meansT) / adj_stds # 평균을 빼고 표준편차로 나눠줌으로써 픽셀 값을 정규화합니다
    return normalized
```

## 9.3 CIFAR-10 파일 전처리

```python
def read_data(directory):
    names = unpickle('{}/batches.meta'.format(directory))['label_names']
    print('names', names)
    data, labels = [], []
    for i in range(1, 6):
        filename = '{}/data_batch_{}'.format(directory, i)
        batch_data = unpickle(filename)
        if len(data) > 0:
            data = np.vstack((data, batch_data['data']))
            labels = np.hstack((labels, batch_data['labels']))
        else:
            data = batch_data['data']
            labels = batch_data['labels']
    
    print(np.shape(data), np.shape(labels))
    data = clean(data)
    data = data.astype(np.float32)
    return names, data, labels
```

```python
names, data, labels = read_data('./cifar-10-batches-py')
```

## 9.5 데이터셋으로부터의 이미지들 시각화하기

```python
import matplotlib.pyplot as plt
import random
```

```python
def show_some_examples(names, data, labels):
    plt.figure()
    rows, cols = 4, 4 # 여러분이 원하는만큼의 행과 열로 바꿀 수 있습니다
    random_idxs = random.sample(range(len(data)), rows * cols) # 데이터셋으로부터 랜덤하게 이미지를 골라 보여줍니다
    for i in range(rows * cols):
        plt.subplot(rows, cols, i + 1)
        j = random_idxs[i]
        plt.title(names[labels[j]])
        img = np.reshape(data[j, :], (24, 24))
        plt.imshow(img, cmap='Greys_r')
        plt.axis('off')
    plt.tight_layout()
    plt.savefig('cifar_examples.png')
```

## 9.6 랜덤 필터 생성과 시각화

```python
import tensorflow as tf
```

```python
W = tf.Variable(tf.random_normal([5, 5, 1, 32])) # 랜덤 필터를 표현할 텐서를 정의합니다
```

```python
def show_weights(W, filename=None):
    plt.figure()
    rows, cols = 4, 8 # 그림 9.4와 같이 32개의 이미지를 보여주기 위한 행과 열을 정의합니다
    for i in range(np.shape(W)[3]): # 각 필터 행렬을 시각화합니다
        img = W[:, :, 0, i]
        plt.subplot(rows, cols, i + 1)
        plt.imshow(img, cmap='Greys_r', interpolation='none')
        plt.axis('off')
    if filename:
        plt.savefig(filename)
    else:
        plt.show()
```

## 9.7 세션을 이용하여 가중치 초기화하기

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    W_val = sess.run(W)
    show_weights(W_val, 'step0_weights.png')
```

## 9.8 컨벌루션 결과 보기

```python
def show_conv_results(data, filename=None):
    plt.figure()
    rows, cols = 4, 8
    for i in range(np.shape(data)[3]):
        img = data[0, :, :, i] # 코드 9.6와는 다르게 이번에는 텐서의 형태가 다릅니다
        plt.subplot(rows, cols, i + 1)
        plt.imshow(img, cmap='Greys_r', interpolation='none')
        plt.axis('off')
    if filename:
        plt.savefig(filename)
    else:
        plt.show()
```

## 9.9 컨벌루션의 시각화

```python
raw_data = data[4, :] # CIFAR-10 데이터셋으로부터 이미지를 가져와서 시각화합니다
raw_img = np.reshape(raw_data, (24, 24))
plt.figure()
plt.imshow(raw_img, cmap='Greys_r')
plt.savefig('input_image.png')

x = tf.reshape(raw_data, shape=[-1, 24, 24, 1]) # 24 x 24 이미지에 대한 입력 텐서를 정의합니다

b = tf.Variable(tf.random_normal([32])) # 필터와 이에 대한 파라미터를 정의합니다
conv = tf.nn.conv2d(x, W, strides=[1, 1, 1, 1], padding='SAME')
conv_with_b = tf.nn.bias_add(conv, b)
conv_out = tf.nn.relu(conv_with_b)

with tf.Session() as sess: # 선택된 이미지에 대해 컨벌루션을 실행합니다
    sess.run(tf.global_variables_initializer())
    
    conv_val = sess.run(conv)
    show_conv_results(conv_val, 'step1_convs.png')
    print(np.shape(conv_val))
    
    conv_out_val = sess.run(conv_out)
    show_conv_results(conv_out_val, 'step2_conv_outs.png')
    print(np.shape(conv_out_val))
```

## 9.11 CNN 가중치 설정

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
names, data, labels = read_data('./cifar-10-batches-py') # 데이터를 로드합니다

x = tf.placeholder(tf.float32, [None, 24 * 24]) # 입력과 출력 플레이스홀더를 정의합니다
y = tf.placeholder(tf.float32, [None, len(names)])

W1 = tf.Variable(tf.random_normal([5, 5, 1, 64])) # 윈도우 크기 5 x 5의 64개 컨벌루션을 적용합니다
b1 = tf.Variable(tf.random_normal([64]))

W2 = tf.Variable(tf.random_normal([5, 5, 64, 64])) # 윈도우 크기 5 x 5의 64개 컨벌루션을 적용합니다
b2 = tf.Variable(tf.random_normal([64]))

W3 = tf.Variable(tf.random_normal([6*6*64, 1024])) # 완전히 연결된 레이어를 도입합니다
b3 = tf.Variable(tf.random_normal([1024]))

W_out = tf.Variable(tf.random_normal([1024, len(names)])) # 완전히 연결된 선형 레이어 관련 변수를 정의합니다
b_out = tf.Variable(tf.random_normal([len(names)]))
```

## 9.12 컨벌루션 레이어 생성

```python
def conv_layer(x, W, b):
    conv = tf.nn.conv2d(x, W, strides=[1, 1, 1, 1], padding='SAME')
    conv_with_b = tf.nn.bias_add(conv, b)
    conv_out = tf.nn.relu(conv_with_b)
    return conv_out
```

## 9.13 맥스 풀링 레이어 생성

```python
def maxpool_layer(conv, k=2):
    return tf.nn.max_pool(conv, ksize=[1, k, k, 1], strides=[1, k, k, 1], padding='SAME')
```

## 9.14 완성된 CNN 모델

```python
def model():
    x_reshaped = tf.reshape(x, shape=[-1, 24, 24, 1])
    
    conv_out1 = conv_layer(x_reshaped, W1, b1) # 첫번째 컨벌루션, 맥스 풀링 레이어를 구성합니다
    maxpool_out1 = maxpool_layer(conv_out1)
    norm1 = tf.nn.lrn(maxpool_out1, 4, bias=1.0, alpha=0.001 / 9.0, beta=0.75)
    
    conv_out2 = conv_layer(norm1, W2, b2) # 두번째 레이어를 구성합니다
    norm2 = tf.nn.lrn(conv_out2, 4, bias=1.0, alpha=0.001 / 9.0, beta=0.75)
    maxpool_out2 = maxpool_layer(norm2)
    
    maxpool_reshaped = tf.reshape(maxpool_out2, [-1, W3.get_shape().as_list()[0]]) # 마지막 부분인 완전히 연결된 레이어를 구성합니다
    local = tf.add(tf.matmul(maxpool_reshaped, W3), b3)
    local_out = tf.nn.relu(local)
    
    out = tf.add(tf.matmul(local_out, W_out), b_out)
    return out 
```

## 9.15 비용과 정확도를 측정하기 위한 연산의 정의

```python
model_op = model()

cost = tf.reduce_mean( # 분류 비용 함수를 정의합니다
    tf.nn.softmax_cross_entropy_with_logits_v2(logits=model_op, labels=y)
)

train_op = tf.train.AdamOptimizer(learning_rate=0.001).minimize(cost) # 비용 함수를 최소화하기 위한 학습 연산을 정의합니다

correct_pred = tf.equal(tf.argmax(model_op, 1), tf.argmax(y, 1))
accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))
```

## 9.16 CIFAR-10 데이터셋을 이용한 신경망 학습

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    onehot_labels = tf.one_hot(labels, len(names), on_value=1., off_value=0., axis=-1)
    onehot_vals = sess.run(onehot_labels)
    batch_size = len(data) // 200
    print('batch size', batch_size)
    for j in range(0, 1000): # 1,000 에포크만큼 반복합니다
        print('EPOCH', j)
        for i in range(0, len(data), batch_size): # 네트워크를 배치로 학습시킵니다
            batch_data = data[i:i+batch_size, :]
            batch_onehot_vals = onehot_vals[i:i+batch_size, :]
            _, accuracy_val = sess.run([train_op, accuracy], feed_dict={x: batch_data, y: batch_onehot_vals})
            if i % 1000 == 0:
                print(i, accuracy_val)
        print('DONE WITH EPOCH')
```
