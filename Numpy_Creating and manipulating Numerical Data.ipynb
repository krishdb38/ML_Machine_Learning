#### Creating arrays

```python
# Manual Construction of arrays
# 1-D
a = np.array([0,1,2,3])
print(a)
print(a.shape)
print(a.ndim)
print(a.min())
print(len(a))
print(a.item)
```

```python
# 2 d , 3 D ...
b = np.array([[0,1,2],
             [3,4,5]])
print(b.ndim)
print(b.shape)
print(len(b))  # Returns the size of the first dimension
```

```python
c = np.array([[[1],[2]],
             [[3],[4]]])
```

```python
print(c.shape)
print(c.ndim)
```

##### Functions for creating arrays

```python
# random Number of evenly spaced
a = np.arange(10) # 0 to 10 numbers 
print(a)

b = np.arange(1,9,2) # start, end, step
# from 1 to 9 , gap 2
print(b)
```

```python
# or Number of Points
c = np.linspace(0,1,6)
# star, end, num points
print(c)

d = np.linspace(0,1,5, endpoint=True)
print(d)
```

```python
# common arrays
a = np.ones([3,3])
print(a)

b = np.zeros([2,2])
print(b)

c = np.eye(3)
print(c)

d = np.diag(np.array([1,2,3,4]))
print(d)

e = np.diag([2,4,6,8]) # Keep these elements as diagnal
print(e)

```

##### Basic Visualization

```python
x = np.array([1,2,3,4,5])
y = np.array([10,8,6,4,2])

plt.plot(x,y)
```

#### 1 D Plotting

```python
x = np.linspace(0,3,20)
y = np.linspace(0,9,20)

a = y.copy()
b = x.copy()

plt.plot(x,y,"red")
plt.plot(x,y,"o")

plt.plot(a,b)
plt.plot(a,b,"o")
```
